```r
library(tidyverse)
```

```r
covidlies = read_csv("./covid_lies.csv")
```

```r
# number of tweets by label

covidlies %>%
    count(label)
```

```r
# number of tweets by misconception_id FOR "pos" label

covidlies %>%
    filter(label == "pos") %>%
    group_by(misconception_id) %>%
    tally()
```

```r
# number of tweets by misconception_id FOR "neg" label

covidlies %>%
    filter(label == "neg") %>%
    group_by(misconception_id) %>%
    tally()
```

```r
# number of tweets by misconception_id FOR "na" label

covidlies %>%
    filter(label == "na") %>%
    group_by(misconception_id) %>%
    tally()
```
